# Convert Runtime column to Float

```python
data_movie = data.copy()
data_movie["Runtime"] = data_movie["Runtime"].str.extract('(\d+)').astype(float)
```

# Convert "Released_Year" Column to Integer.

```python
data_movie["Released_Year"] = pd.to_numeric(data_movie["Released_Year"],
                                            errors="coerce", downcast = "integer")
```

```python
print(data_movie[data_movie["Released_Year"].isnull()])
```

# Replace NaN Released Year for film "Apollo 13" with "1995".

```python
data_movie.dropna(subset=["Released_Year"], inplace=True)
# By using "dropna" in the Released Year the All NaN value Should be drop Where 
# the value is "True".
```

```python
data_movie["Released_Year"] = data_movie["Released_Year"].fillna(1995)
# By using "fillna()" the replace NaN value will filled with the value
# which given in the parameter.
```

# Replace Missing values with "Mean".

```python
mean_meta_score = data_movie["Meta_score"].mean()
data_movie.fillna({"Meta_score" : mean_meta_score}, inplace = True)
```

# Replace the NaN with the "Unknown" Category.


```python
data_movie.fillna({"Certificate" : "Unknown"}, inplace=True)
```

# Convert Gross Column to Float and Replace NaN value with "median"

```python
data_movie["Gross"] = data_movie["Gross"].str.replace(',', '').astype(float)
```

```python
median_gross = data_movie["Gross"].median()
data_movie.fillna({"Gross" : median_gross}, inplace=True)
```

# By the help of Matplotlib.pyplot & Seaborn Libraries shows the Histogram for Overall Vote Distribution.

```python
data_movie.columns.tolist()
```

```python
votes_by_year = data_movie.groupby('Released_Year')["No_of_Votes"].sum().reset_index()

plt.figure(figsize=(20, 5))
sns.lineplot(data=votes_by_year, x="Released_Year", y="No_of_Votes")
plt.title("Voting Trends Over the Years")
plt.xlabel("Released Year")
plt.ylabel("Sum of the Votes")
plt.show()
```

# The Highest Average  of Gross

```python
mean_gross_by_year = data_movie.groupby("Released_Year")["Gross"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=mean_gross_by_year.index, y=mean_gross_by_year.values)
plt.title("Gross value per year")
plt.xlabel("Released Year")
plt.ylabel("Mean of the Gross")
plt.xticks(rotation=90) # Rotate x-axis labels for better readability.
plt.show()
```

# Display Top 10 Highest Score Movie Titles and its Directors

```python
Top_10_movies_Directors = data_movie.nlargest(10, "Meta_score")[["Series_Title",
                                                                 "Meta_score","Director"]].set_index("Series_Title")
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Meta_score", y=Top_10_movies_Directors.index,
            hue="Director", data=Top_10_movies_Directors, dodge=False)
plt.title(" Top 10 Highest Score Movie Titles and its Directors")
plt.xlabel("Meta_score")
plt.ylabel("Movie Title")
plt.legend(title="Director", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
